````python
def PatternCount(Text, Pattern):
    count = 0
    for i in range(len(Text)-len(Pattern)):
        if Text[i:(i+len(Pattern))] == Pattern:
            count = count + 1
    return(count)
````

````python
Text="Hola, ¿como estas?. Hola, bien gracias. Y tú, ¿qué cuentas?"
Pattern="as"
PatternCount(Text, Pattern)
````
