¿Cómo conectarse al servidor Linux?
----------------

Nuestra universidad dispone de un servidor el cual puede ser accedido a través de la siguiente dirección IP:

```
146.83.191.171
```

Para conectarse a esta dirección se debe usar la herramienta [SSH](https://en.wikipedia.org/wiki/Secure_Shell). Si estás trabajando desde:

* Linux o Mac, ya tienes SSH instalado.
* Windows, necesitas instalar [Cygwin](https://www.cygwin.com/).

Para conectarte al servidor (con Cygwin si usas Windows) escribe remplazando username con tu nombre de usuario (ver más abajo cual es tu nombre de usuario):

```
ssh username@146.83.191.171 -p 60222
```

Si una ventana se abre, selecciona yes. Despues de esto, ya estás en el servidor Linux.

Familarización con Linux
----------------

Comandos comunes

```
$ mkdir testDir # creates a directory called testDir
$ cd testDir    # moves into directory testDir
$ pwd           # outputs present working directory
$ cd -          # moves to the previous directory
$ rm testDir    # remove a file, since testDir is a directory rather than a file, it will complain
$ rm -r testDir # it will remove the folder along with its contents
```

Creemos un archivo de pruena. Para ésto utilizaremos el editor ```vi```:

```
$ mkdir testDir
$ cd testDir
$ vi testFile.txt
```

Verifiquemos si el archivo ha sido guardado:

```
$ ls
testFile.txt
```
